# Importing and Sample Data

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic Plot
plt.plot(x, y)
plt.show()
```

# Line Plot

```python
# Basic Line Plot
plt.plot(x, y, label='Sine Wave')

# Add title and labels
plt.title('Sine Wave Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

# Scatter Plot

```python
# Sample data
x = np.random.rand(50)
y = np.random.rand(50)

# Basic Scatter Plot
plt.scatter(x, y, color='blue', marker='o')

# Add title and labels
plt.title('Scatter Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Display the plot
plt.show()
```

# Bar Plot

```python
# Sample data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 4]

# Basic Bar Plot
plt.bar(categories, values, color='green')

# Add title and labels
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Display the plot
plt.show()
```

# Horizontal Bar Plot

```python
# Basic Horizontal Bar Plot
plt.barh(categories, values, color='orange')

# Add title and labels
plt.title('Horizontal Bar Plot Example')
plt.xlabel('Values')
plt.ylabel('Categories')

# Display the plot
plt.show()
```

# Histogram

```python
# Sample data (normally distributed)
data = np.random.randn(1000)

# Basic Histogram
plt.hist(data, bins=30, color='purple')

# Add title and labels
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

# Pie Chart

```python
# Sample data
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']

# Basic Pie Chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```

# Box Plot

```python
# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Basic Box Plot
plt.boxplot(data, vert=True, patch_artist=True)

# Add title and labels
plt.title('Box Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')

# Display the plot
plt.show()
```

# Subplots

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 1 row and 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# First subplot (Sine)
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Wave')

# Second subplot (Cosine)
ax2.plot(x, y2, color='red')
ax2.set_title('Cosine Wave')

# Display the plots
plt.show()
```

# Customizing Plots

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Customize plot (linewidth, color, linestyle)
plt.plot(x, y, linewidth=2, color='red', linestyle='--')

# Add grid
plt.grid(True)

# Add title and labels
plt.title('Customized Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Display the plot
plt.show()
```

# Logarithmic Scale Plot

```python
# Sample data
x = np.linspace(1, 100, 100)
y = np.log(x)

# Logarithmic scale plot
plt.plot(x, y)

# Set logarithmic scale for y-axis
plt.yscale('log')

# Add title and labels
plt.title('Logarithmic Scale Example')
plt.xlabel('X Axis')
plt.ylabel('Log(Y)')

# Display the plot
plt.show()
```

# Saving a Plot

```python
# Sample plot
plt.plot(x, y)

# Save the plot as a file
plt.savefig('plot.png')

# To save without white margins
plt.savefig('plot.png', bbox_inches='tight')

# Display the plot
plt.show()
```
