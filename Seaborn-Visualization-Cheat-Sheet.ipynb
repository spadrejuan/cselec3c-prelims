# Import and Sample Data


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load sample dataset from Seaborn
tips = sns.load_dataset('tips')

# Basic plot with Seaborn
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.show()
```

# Scatter Plot


```python
# Basic Scatter Plot
sns.scatterplot(data=tips, x='total_bill', y='tip')

# Add title and labels
plt.title('Scatter Plot: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Display the plot
plt.show()
```

# Line Plot


```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic Line Plot
sns.lineplot(x=x, y=y)

# Add title and labels
plt.title('Line Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Display the plot
plt.show()
```

# Bar Plot


```python
# Bar plot with 'day' as categories and mean of 'total_bill'
sns.barplot(data=tips, x='day', y='total_bill')

# Add title
plt.title('Bar Plot: Average Total Bill by Day')

# Display the plot
plt.show()
```

# Box Plot


```python
# Box plot of total_bill by day
sns.boxplot(data=tips, x='day', y='total_bill')

# Add title
plt.title('Box Plot: Total Bill by Day')

# Display the plot
plt.show()
```

# Violin Plot


```python
# Violin plot (combines box plot and KDE)
sns.violinplot(data=tips, x='day', y='total_bill')

# Add title
plt.title('Violin Plot: Total Bill by Day')

# Display the plot
plt.show()
```

# Histogram / Distribution Plot


```python
# Basic Histogram (distribution plot) of 'total_bill'
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)

# Add title
plt.title('Histogram: Total Bill Distribution')

# Display the plot
plt.show()
```

# Pair Plot


```python
# Pair plot (scatterplot matrix) with different relationships
sns.pairplot(tips)

# Display the plot
plt.show()
```

# Heatmap


```python
# Select only numeric columns
numeric_tips = tips.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numeric_tips.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Display the plot
plt.title('Heatmap: Correlation Matrix (Numeric Only)')
plt.show()
```

# Facet Grid


```python
# FacetGrid to visualize tips by time and day with different hues for smokers
g = sns.FacetGrid(tips, col="time", row="smoker", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip")

# Display the plot
plt.show()
```

# Count Plot


```python
# Count plot for categorical data ('day')
sns.countplot(data=tips, x='day')

# Add title
plt.title('Count Plot: Number of Orders per Day')

# Display the plot
plt.show()
```

# Kernel Density Estimation (KDE) Plot


```python
# KDE plot (smooth distribution of 'total_bill')
sns.kdeplot(data=tips, x='total_bill', fill=True)

# Add title
plt.title('KDE Plot: Total Bill')

# Display the plot
plt.show()
```

# Joint Plot


```python
# Joint plot showing scatter plot with marginal histograms for total_bill vs tip
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')

# Display the plot
plt.show()
```

# Swarm Plot


```python
# Swarm plot of total_bill by day
sns.swarmplot(data=tips, x='day', y='total_bill')

# Add title
plt.title('Swarm Plot: Total Bill by Day')

# Display the plot
plt.show()
```

# Styling with Seaborn

```python
# Set the Seaborn theme
sns.set_theme(style='darkgrid')

# Basic Scatter Plot
sns.scatterplot(data=tips, x='total_bill', y='tip')

# Add title
plt.title('Styled Scatter Plot: Total Bill vs Tip')

# Display the plot
plt.show()
```

# Regression Plot

```python
# Basic regression plot with linear trend line
sns.regplot(data=tips, x='total_bill', y='tip')

# Add title
plt.title('Regression Plot: Total Bill vs Tip')

# Display the plot
plt.show()
```

# Saving a Plot

```python
# Sample plot
sns.scatterplot(data=tips, x='total_bill', y='tip')

# Save the plot to a file
plt.savefig('seaborn_plot.png')

# Display the plot
plt.show()
```
